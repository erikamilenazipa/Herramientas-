```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
df = pd.read_csv("Eventos_de_inter_s_en_salud_p_blica_registrados_en_el_Municipio_de_Acac_as_20250505.csv")
```

```python

valores_cantidad = df.Edad.unique()


plt.xticks(valores_cantidad)


formatter = FuncFormatter(lambda x, pos: f'{int(x):,}'.replace(',', '.'))


plt.gca().yaxis.set_major_formatter(formatter)


plt.title('Histograma edades en que se contrajeron la enfermedad')
plt.xlabel('tipo de enfermedad')
plt.ylabel('Frecuencia')

df = df[~df.Edad.isin([1,2])]



n, bins, patches = plt.hist(df.Edad, color="Red")

for i in range(len(n)):
    plt.text(bins[i] + 0.5, n[i] + 0.1, f'{int(n[i]):,}'.replace(',', '.'), ha='center', va='bottom', fontsize=10)

plt.show()
```
