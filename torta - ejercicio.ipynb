```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Eventos_de_inter_s_en_salud_p_blica_registrados_en_el_Municipio_de_Acac_as_20250505.csv")
```

```python
df.shape
```

```python
tabla_muertes = df.groupby('Area')['Edad'].sum()


plt.title('Distribución del area por edad')


plt.pie(tabla_muertes, labels=tabla_muertes.index, autopct='%1.1f%%', startangle=180, colors=['red', 'blue', 'black'])


plt.legend()


plt.show()
```

```python

tabla_muertes = df.groupby('Nombre del Evento')['Edad'].sum()

plt.title('Distribución de la enfermedad por area')


plt.pie(tabla_muertes, labels=tabla_muertes.index, autopct='%1.1f%%', startangle=180, colors=['red', 'blue', 'black'])


plt.legend()

```
